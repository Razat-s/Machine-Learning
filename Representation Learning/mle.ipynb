```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

### Generated samples are generated with some parameters and for estimated samples, parameters were estimated using MLE.

```python
N = 1000 # no. of samples
```

```python
# Binomial Distribution
p = 0.6 # parameter
trials  = 1

binom  = np.random.binomial(trials,p,N) # generating samples

#estimating parameter p
p_mle = np.mean(binom)

gen_binom = np.random.binomial(trials,p_mle,N)

#plotting generated samples
a = plt.subplot(1,2,1)
plt.hist(binom, bins = 100)
a.set_title('generated Samples')

#plotting estimated samples
b = plt.subplot(1,2,2)
plt.hist(gen_binom, bins = 100)
b.set_title('Estimated Samples')
plt.show()
```

```python
# Poisson Distribution
lamda = 30
poisson  = np.random.poisson(lamda,N) # generating samples

#estimating parameter lamda
lamda_mle = np.mean(poisson)

gen_poisson = np.random.poisson(lamda_mle,N)

#plotting generated samples
a = plt.subplot(1,2,1)
plt.hist(poisson, bins = 100)
a.set_title('generated Samples')

#plotting estimated samples
b = plt.subplot(1,2,2)
plt.hist(gen_poisson, bins = 100)
b.set_title('Estimated Samples')
plt.show()
```

```python
# Exponential Distribution
lamda = 3
exp  = np.random.exponential(lamda,N) # generating samples

#estimating parameter lamda
lamda_mle = 1/np.mean(exp)

gen_exp = np.random.exponential(lamda_mle,N)

#plotting generated samples
a = plt.subplot(1,2,1)
plt.hist(exp, bins = 100)
a.set_title('generated Samples')

#plotting estimated samples
b = plt.subplot(1,2,2)
plt.hist(gen_exp, bins = 100)
b.set_title('Estimated Samples')
plt.show()
```

```python
# Gaussian Distribution
mu = 0 # parameter
sigma  = 1

gauss  = np.random.normal(mu,sigma,N) # generating samples

#estimating parameter mu,sigma
mu_mle = np.mean(gauss)
sigma_mle = np.std(gauss)

gen_gauss = np.random.normal(mu_mle,sigma_mle,N)

#plotting generated samples
a = plt.subplot(1,2,1)
plt.hist(gauss, bins = 100)
a.set_title('generated Samples')

#plotting estimated samples
b = plt.subplot(1,2,2)
plt.hist(gen_gauss, bins = 100)
b.set_title('Estimated Samples')
plt.show()
```

```python
# Laplacian Distribution
mu = 0 # parameter
b  = 1

lap = np.random.laplace(mu,b,N) # generating samples

#estimating parameter mu,b
mu_mle = np.median(lap)
b_mle = np.mean(np.abs(lap - mu))

gen_lap = np.random.laplace(mu_mle,b_mle,N)

#plotting generated samples
a = plt.subplot(1,2,1)
plt.hist(lap, bins = 100)
a.set_title('generated Samples')

#plotting estimated samples
b = plt.subplot(1,2,2)
plt.hist(gen_lap, bins = 100)
b.set_title('Estimated Samples')
plt.show()
```
